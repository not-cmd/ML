```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/Users/divyeshmedidi/Documents/python programming/ML intro/py/ML/2_linear_reg_multivariate/homeprices.csv")
df.head()
```

```python
import math
median_floors = math.floor(df.bedrooms.median())
median_floors
```

```python
df.bedrooms=df.bedrooms.fillna(median_floors)
df
```

```python
df.head(10)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']],df.price)
```

```python
reg.coef_
```

```python
reg.predict([[3000,3,40]])
```
